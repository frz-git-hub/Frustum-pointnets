```python
%%writefile /content/frustum-pointnets/dataset/KITTI/object/sample.py

import os

def main():
  BASE_DIR = os.path.dirname(os.path.abspath(__file__))
  
  train_path = os.path.join(BASE_DIR, 'training')
  test_path = os.path.join(BASE_DIR, 'testing')

  train_image_path = os.path.join(train_path, 'image_2')
  train_point_path = os.path.join(train_path, 'velodyne')
  train_label_path = os.path.join(train_path, 'label_2')
  train_calib_path = os.path.join(train_path, 'calib')

  test_image_path = os.path.join(test_path, 'image_2')
  test_point_path = os.path.join(test_path, 'velodyne')
  test_calib_path = os.path.join(test_path, 'calib')

  if os.path.isdir('sample_data'):
    os.system('rm -r sample_data/')

  os.mkdir('sample_data')
  
  os.mkdir('sample_data/training')
  os.mkdir('sample_data/testing')

  os.mkdir('sample_data/training/image_2')
  os.mkdir('sample_data/training/velodyne')
  os.mkdir('sample_data/training/label_2')
  os.mkdir('sample_data/training/calib')

  os.mkdir('sample_data/testing/image_2')
  os.mkdir('sample_data/testing/velodyne')
  os.mkdir('sample_data/testing/calib')

  sample_num = 20
  for i in range(20) :

    # training
    src = os.path.join(train_image_path, '%06d.png' % i)
    cmd = "cp {} {}".format(src,"sample_data/training/image_2")
    os.system(cmd)

    src = os.path.join(train_point_path, '%06d.bin' % i)
    cmd = "cp {} {}".format(src,"sample_data/training/velodyne")
    os.system(cmd)

    src = os.path.join(train_label_path, '%06d.txt' % i)
    cmd = "cp {} {}".format(src,"sample_data/training/label_2")
    os.system(cmd)

    src = os.path.join(train_calib_path, '%06d.txt' % i)
    cmd = "cp {} {}".format(src,"sample_data/training/calib")
    os.system(cmd)

    # testing
    src = os.path.join(test_image_path, '%06d.png' % i)
    cmd = "cp {} {}".format(src,"sample_data/testing/image_2")
    os.system(cmd)

    src = os.path.join(test_point_path, '%06d.bin' % i)
    cmd = "cp {} {}".format(src,"sample_data/testing/velodyne")
    os.system(cmd)

    src = os.path.join(test_calib_path, '%06d.txt' % i)
    cmd = "cp {} {}".format(src,"sample_data/testing/calib")
    os.system(cmd)
  
  os.system('zip -r sample_data.zip sample_data')
  os.system('rm -r sample_data/')

if __name__ == "__main__":
  main()
  print('ok')
```
